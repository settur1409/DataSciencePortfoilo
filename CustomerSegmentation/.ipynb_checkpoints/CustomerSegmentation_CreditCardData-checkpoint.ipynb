*Reading and previewing Data*
*There are missing values and will be treated using fillna()*


```python
CC_Data = pd.read_csv("CC GENERAL.csv")
CC_Data.head()
```

*Filling missing values with mean of the attribute as the column type is float*
*Viewing the percentage of missing values to the sum of elements*

```python
CC_Data['MINIMUM_PAYMENTS'] = CC_Data['MINIMUM_PAYMENTS'].fillna(np.mean(CC_Data.MINIMUM_PAYMENTS))
CC_Data['CREDIT_LIMIT'] = CC_Data['CREDIT_LIMIT'].fillna(np.mean(CC_Data['CREDIT_LIMIT']))
CC_Data.isna().sum()/len(CC_Data) *100
```

### Features for evaluation :: [purchase transactions, credit limit, purchase frequency , Balance, tenure]

```python
AnalysisData = CC_Data[['PURCHASES_TRX', 'PURCHASES_FREQUENCY', 'PURCHASES', 'BALANCE','CREDIT_LIMIT','TENURE']]
plt.figure(figsize=(20,15))
sns.pairplot(AnalysisData, hue= 'TENURE')
plt.savefig("DataSummarized with only required attributes for model building.jpg")
plt.show()
```

```python
#Evaluating segmentation of PURCHASES_TRX with BALANCE
from sklearn.cluster import KMeans
from scipy.spatial import distance

X1 = AnalysisData[['PURCHASES_TRX', 'BALANCE']]
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(X1)
    distortions.append(sum(np.min(distance.cdist(X1, kmeanModel.cluster_centers_, 'euclidean'), axis=1)) / X1.shape[0])
    
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show() 
```

```python
kmeanModel = KMeans(n_clusters=5).fit(X1)

plt.scatter(X1.iloc[:,0], X1.iloc[:,1], c=kmeanModel.labels_)
plt.xlabel("PURCHASES TRANSACTIONS")
plt.ylabel("BALANCE")
plt.scatter(kmeanModel.cluster_centers_[:,0] ,kmeanModel.cluster_centers_[:,1], color='black')
plt.savefig("Segmenting Customers Based on PURCHASES TRANSACTIONS with BALANCE.jpg")
```
